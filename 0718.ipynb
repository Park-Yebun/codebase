numeric type
---
#### 1.  정수자료형(int)

- 진수표현
    
    - 파이썬에서 기본 숫자입력은 10진수. 따라서 int() 함수는 여러 진법의 수를 10진수로 변환하는 것!

0b10 
0o30
0x10

bin()   정수인 매개변수를 2진수로 변환해주는 함수

oct()   정수인 매개변수를 8진수로 변환해주는 함수

hex()   정수인 매개변수를 16진수로 변환해주는 함수




 
#### 2.  실수자료형(float)

- 유한 정밀도 : 실수 자료형을 컴퓨터에 표현한 값은 실수 전체가 아니라 한정된 데이터값임(소수점 n째 자리 데이터의 유실), 제한된 데이터로 인해 근삿값을 표현한 것=실수
- 위의 유한 정밀도로 인해 근삿값으로 실수 자료형을 연산할 경우 원하는 값이 나오지 않는 등의 문제 발생(floating point rounding error) 

-> math 모듈을 활용해 계산하거나 임의의 작은 수를 기준으로 하여 출력되게끔 함


sequence types(text sequence type)
---

#### 1.  문자열(str)

- 각각의 문자로 이루어진 배열, 순서가 있고 변경불가능 한덩어리로 할당됨
- **escape sequence** : 파이썬의 일반적인 문법규칙을 벗어나게 해주는 특정 문자들의 조합 (＼n, ＼t, ＼＼, ＼', ＼")
- **syting interpolation** : **f-string**을 이용해 문자열 내에 변수/표현식을 삽입.

- 문자열 내에서 할 수 있는 것들 : index[], slicing[:], len() -> 다만 문자열 자료형은 변경불가능하므로 슬라이싱을 통해 추출할 경우 원래 값을 수정하는 것이 아닌 완전히 새로운 주소로 복사하는 것!


#### 2.  리스트(list)
- 문자들이 모여 만들어진 그룹. 순서가 있고 변경도 가능. 쉼표를 기준으로 각각의 단어로 할당됨. **리스트=대괄호[], 튜플=소괄호()**
- 문자열 내에서 할 수 있는 것들 : index[], slicing[:], len() -> 다만 문자열 자료형은 변경불가능하므로 슬라이싱을 통해 추출할 경우 원래 값을 수정하는 것이 아닌 완전히 새로운 주소로 복사하는 것!
- 리스트는 가변형이므로 인덱스를 이용해 특정 값을 변경하거나 반복하거나 없앨수 있음.


#### 3.  튜플(tuple)
- 리스트랑 성질이 똑같으나 변경 불가능, 값을 반드시 0개 이상 가져야함 --튜플의 요소가 하나라면 반드시 쉼표를 포함할 것

#### 4.  레인지(range)
- 정수의 연속된 범위값 지정, 변경 불가능


Non-sequence Types
---
#### 1.  딕셔너리(dictionary)
- key와 value가 한쌍, **순서와 중복이 없음, 변경 가능**
- key 값은 불변하는 값만 사용 가능(리스트 안됨), value는 모든 자료형 가능

#### 2.  세트(set)
- 순서도 없고, 중복도 없음, 변경 가능
- 수학에서의 집합과 같은 성질 (합집합| 차집합- 교집합&)

other type
---
#### 1. none
#### 2. boolean
- True or False

콜렉션
---
- 자료형 중 여러개의 요소를 담을 수 있는 것들/numeric, range, set 제외 -- numeric type은 int함수 없이는 기본 문자형으로 저장되기때문에 실제로는 한가지 요소만 담는다.

형변환
---
1. 암시적 형변환
    - boolean 과 numeric type에서만 가능. 암시적 형변환이 일어날때는 더 범위가 넓은 형의 구조를 따른다.

2. 명시적 형변환
    - 모든 값은 딕셔너리와 레인지로 형변환이 불가함
    - 딕셔너리는 시퀀스형으로 변환할때 키 값만 형변환이 가능(str 제외)
    - 컬렉션간 형변환에서 딕셔너리를 다른 형으로 변환하는 경우 value값은 버려짐

연산자(operator)
--- 
- 산술 연산자와 복합 연산자
    - 산술 연산자 : 기본 연산자
    - 복합 연산자 : 연산과 할당이 같이 이루어지는 연산자 (+=, +-, +/, +*등)
- 비교 연산자 
    - is 비교 연산자 : 식별성 비교 True/False
    - == 와 is 의 차이점 : ==은 값을 비교, is는 참조하는 주소를 비교(주소값이 같냐?)
- 논리연산자
    - and는 논리곱 True and True => 1 * 1 => 1(True)
    - or은 논리합 True or False => 1 + 0 => 1(True)
    - not은 논리부정 값의 반대로 할당
    - **단축평가** : 오른쪽값을 평가하지 않고 결과를 결정 --불필요한 연산을 피함으로써 코드 실행을 최적화
- 멤버십 연산자
    - in/not in : 왼쪽 피연산자가 오른쪽 시퀀스에 속하는지?
- 시퀀스형 연산자
    - +, -




```python
# 진법 변경 (10진수 > n진수)
print(bin(12))
print(oct(12))
print(hex(12))

# 지수 표현
print(314e-2) #314에 -(10의 2제곱)을 곱한 것
print(314e2) 

# f-string 응용
greeting = 'hi'
print(f'{greeting}:^10') #10칸씩 띄고 가운데
print(f'{greeting}:>10') #10칸 띄고 오른쪽 정렬
print(f'{greeting}:.')
```
