### 알고리즘 복잡도 분석
1. 시간 복잡도의 점근적 표기법 : 시간복잡도는 입력크기인 N에 영향을 받은 함수형태로 표기할 수 있다. 점근적 표기란 N이 계속 커질 때 함수가 어떤 크기로 근사하게 되는지를 분석하는 방식이다.
  - O(Big-Oh)- 표기 : 복잡도를 나타내는 함수의 상한 근사값, 단순화된 함수에 임의의 상수 c를 곱한값은 기존 함수와 비교했을때 어느 교차점 이상을 지나면 무조건 기존 함수보다 클 수 밖에 없다. >> 최대 이만큼의 시간이 걸린다.
  - Ω(Big-Omega)- 표기 : 복잡도를 나타내는 함수의 하한 근사값, 임의의 상수 c를 1로 놓았을때 일정 교차점 이후부터는 기존 함수보다 크기가 작아진다. >> 최소 이만한 시간은 걸린다.
  - θ(Theta)- 표기 : O-표기와 Ω-표기가 동일한 경우에만 사용, 동일한 증가율을 가지지 않는 n, nlogn, n^3, 2^n은 θ-표기로 나타낼 수 없다.

### 비트 연산자
1. &: AND 연산, 비트가 둘다 1이여야 결과값이 1이됨 >> 특정 비트를 1에서 0으로 만들때, 연산할 비트에서 해당 자리만 0으로 해주면 됨. 나머지는 1:1 = 1, 0:1 = 0으로 변화없음
2. |: OR연산, 비트가 둘 중 하나가 1이면 결과값이 1임 >> 특정 비트를 0에서 1로 만들때 사용, 연산 비트의 해당 자리만 1로 써주기.
3. ^: exclusive or(배타적인 or연산자), 이름뜻 그대로 두 비트가 서로 달라야 결과값이 1이됨 >> 특정 비트만 반전시키고 싶다면 연산비트의 동일 자리의 값만 1로 써주기
4. ~ : invert, '거꾸로하다' 라는 이름 그대로 전체 비트 값을 반전시키는 연산자
- ** 비트 연산자들은 비트와 비트끼리도 연산가능하지만 (컴퓨터가 10진수를 2진수 형태로 저장해놓기 때문에) 10진수와 혼용해서 사용이 가능하다!**

* 3번과 5번자리의 값이 1인(10진수로는 20)비트를 연산자 표현식으로 만드는법 : 1<<5 | 1<<3 해준다. or는 둘 중 하나만 1이면 결과값이 1이니까 3번 5번자리가 1인 비트가 완성된다.

- 데이터배열 및 처리방식
1. 빅 엔디안 : 큰 단위부터 전송 ex)12, 34 ,,,
2. 리틀 엔디안 : 작은 단위부터 전송, 일반적인 데스크탑 컴퓨터에 탑재 ex)34, 12 ,,,

- 비트 연산자를 이용해 데이터의 순서 뒤집는 방법
  - i만큼 쉬프트해 준 비트를 0xff를 이용해 마스킹(해당 자리의 비트값을 제외하고 0으로 처리)해준것을 & 연산하여 출력하면 해당 데이터 먼저 출력됨

### 진수
- 10진수를 타진수로 변환할 때 : 일반적인 계산방법은 10진수를 원하는 타진법의 수로 나눈 나머지를 거꾸로 읽는 방식
- 2진법과 8진법간 변환 : 2진법->8진법은 3자리씩 비트를 묶어주기, 8진법->2진법은 3자리씩 비트를 나열하기
- 2진법과 16진법간 변환 : 2진법->16진법은 4자리씩 비트를 묶어주기, 16진법->2진법은 4자리씩 비트를 나열하기

1. 2진법으로 음의정수 표현
  - 1의 보수(one's complement) : 절댓값을 비트로 나타낸다음 최상위 비트를 1로 표현해준다. 부호를 제외한 나머지값을 반전시켜준다. -> 0도 양수와 음수로 나뉜다는 단점
  - 2의 보수(two's complement) : 1의 보수에서 최하위 비트에 1을 더해준다. -> 숫자끼리의 비트연산도 가능하다는 장점! -6의 2의 보수 표현에 10을 더해주면 4의 2진수가 결과값으로 나옴
2. 2진법으로 실수 표현
  - 부동 소수점 표기법 : 소수점의 위치를 왼쪽의 가장 유효한 위치로 고정시키고(정수부의 첫번째 자리를 1로 만들고) 변화된 자릿수만큼 밑수의 지수승을 곱해준다.
  - 실수의 저장
    1. 단정도(부호 1비트/지수 8비트/가수 23비트) 
    2. 배정도(부호 1비트/지수 11비트/가수 52비트)
    - 부동 소수점으로 표기된 부분에서 소수점 이하자리만큼 가수부에 넣어주고 빈자리는 0으로 채워줌, 지수부분을 2진법 변환하여 지수부에 넣어줌
    - 다만 단정도는 지수부분에 8비트만큼밖에 할당이 되지 않으므로 **익세스 표현법**을 통해 -256부터 256까지 8비트의 두배만큼의 데이터양을 단정도로 표현해줄 수 있다.
